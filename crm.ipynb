```python
def handler_data(execfile):
    df = pd.read_excel(execfile)
    df.columns.to_list()
    df['设备编号\n（无需填写，编号由本系统自动生成）'] = ''
    df['设备分类\n（请前往系统内左侧菜单栏：基础设置→设备类别查询后填入）'] = '电脑设备'
    # 设备类别处理
    def sblb(row):
        if 'YPZ' in row['固定资产编号']:
            return '有盘站'
        if 'G' in row['固定资产编号']:
            return '便携机'
        if 'PA' in row['固定资产编号']:
            return '平板'
    df['设备类别\n（请前往系统内左侧菜单栏：基础设置→设备类别查询后填入）'] = df.apply(sblb,axis=1)

    # 处理设备品牌
    def sbpp(row):
        if 'hp' in row['牌子型号']:
            return '惠普'
        if 'HP' in row['牌子型号']:
            return '惠普'
        if '威' in row['牌子型号']:
            return '威讯升腾'
        if 'MATE' in row['牌子型号']:
            return '华为'
        
    df['设备品牌'] = df.apply(sbpp,axis=1)

    # 处理设备型号
    def sbxh(row):
        if 'hp' in row['牌子型号']:
            return row['牌子型号'].split('hp')[1]
        if 'HP' in row['牌子型号']:
            return row['牌子型号'].split('HP')[1]
        if '威' in row['牌子型号']:
            return 'VDH510'
        if 'MATE' in row['牌子型号']:
            return 'MATEPAD PRO'
        
    df['设备型号'] = df.apply(sbxh,axis=1)

    df.rename(columns={'配置': '设备配置'}, inplace=True)
    df['当前管理状态'] = '在用'

    def zrr(row):
        # for key,value in email_name.items():
        return email_name[row['使用人']]
    df['责任人\n（填写OA账号）'] = df.apply(zrr,axis=1)
    df['使用人\n（不填写默认责任人）'] = ''
    df['使用部门\n（请前往系统内左侧菜单栏：系统设置→组织架构查询后填入）'] = shiybm
    df['使用内设部门\n（请前往系统内左侧菜单栏：系统设置→组织架构查询后填入）'] = df['位置']
    df['数量'] = 1
    df['单位'] = '台'
    df.rename(columns={'单价': '单价\n(单位元，只填数额)'}, inplace=True)
    df['增加方式'] = '购入'
    df.rename(columns={'位置': '详细地点'}, inplace=True)
    df['增加方式'] = '购入'
    df['增加方式'] = '购入'
    df['设备状态'] = '使用中'

    def format_mac_address(mac):
        return '-'.join(mac[i:i+2] for i in range(0, len(mac), 2))

    def blue_tooth_ip(row):
        if '蓝牙' in row['备注']:
            return row['MAC地址']
        return ''
        
        
    df['MAC地址\n（格式：XX-XX-XX-XX-XX-XX,\n多个使用；号隔开）'] = df['MAC地址'].apply(format_mac_address,axis=1)
    df['蓝牙Mac\n（格式：XX-XX-XX-XX-XX-XX,\n多个使用；号隔开）']= df['MAC地址'].apply(blue_tooth_ip,axis=1)
    df['是否信创'] = df['设备品牌'].apply(lambda x:'是' if '威迅' in x else '否'  )
    df.rename(columns={'IP地址': 'IP\n（格式：XXX.XXX.XXX.XXX,\n多个使用；号隔开）'}, inplace=True)

    df['IP地址'] = df['IP地址'].fillna('0.0.0.0')
    df.rename(columns={'是否监控': '监控状态）'}, inplace=True)
    df['使用人员属性'] = df['使用人\n（不填写默认责任人）'].apply(lambda x: '员工' if x in user_arrt['员工'] else '经纪人')
    df.rename(columns={'固定资产编号': '设备序列号(SN)'}, inplace=True)
    df['资产来源'] = '购入'
    df.rename(columns={'维护日期': '状态变动日期'}, inplace=True)
    def format_date(date_str):
        if len(date_str) == 8:
            return f"{date_str[:4]}-{date_str[4:6]}-{date_str[6:]}"
        return date_str
    df['状态变动日期'] = df['状态变动日期'].apply(format_date)
    df['IMEI/ESN/MEID'] = ''
    df['入库日期\n(格式：YYYY-MM-DD)'] = df['入库日期\n(格式：YYYY-MM-DD)'].apply(format_date)
    df['购买日期\n(格式：YYYY-MM-DD)'] = df['购买日期\n(格式：YYYY-MM-DD)'].apply(format_date)
    df['领用日期\n(格式：YYYY-MM-DD)'] = df['领用日期\n(格式：YYYY-MM-DD)'].apply(format_date)
    df.rename(columns={'用途': '设备用途'}, inplace=True)
    df.rename(columns={'备注': '备注\n(建议填写旧系统编号'}, inplace=True)
    df['项目编号'] = ''
    df['项目名称'] = ''
    df['使用年限\n(单位年，仅填数字)'] = 2025-int(df['购买日期\n(格式：YYYY-MM-DD)'].str[:4])
    df['成本中心']= ''
    df['预算承担部门'] = shiybm
    df['税率(%)\n(只填数额，不用带%号)'] = 0
    df['原值\n(单位元，只填数额)'] = 0
    df['净值\n(单位元，只填数额)'] = 0
    df['净残值比例(%)\n(只填数额，不用带%号)'] = 0
    df['年折旧率(%)\n(只填数额，不用带%号)'] = 0
    df['EAS编号'] = ''
    df['CRM编号'] = df['固定资产编号']


    # zs = zs[new_columns]
    df.head()
    
```
